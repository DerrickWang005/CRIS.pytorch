```python
from PIL import Image

import matplotlib.pyplot as plt
from single_img_inference import get_inferer
```

```python
def plot(img, pred, txt, savefig=False):
    fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, sharey=True, figsize=(10, 5))
    
    ax1.imshow(img)
    ax1.set_title("Image")

    ax2.imshow(pred, cmap="gray")
    ax2.set_title(txt)

    if savefig:
        fig.savefig(f"{txt}.png", bbox_inches='tight', pad_inches=0)
```

```python
img_path="/mnt/Enterprise/PUBLIC_DATASETS/camus_database_png/training/image/patient0001_2CH_ED.png"
```

```python
with Image.open(img_path) as img:
    plt.imshow(img, cmap="gray")
```

```python
infer = get_inferer("/mnt/Enterprise/miccai_2023_CRIS/CRIS.pytorch/exp_old/CRIS_R50/best_model.pth")
```

```python
img, txt, pred = infer(img_path, "left ventricular cavity in the heart ultrasound image")
plot(img, pred, txt, savefig=True)
```

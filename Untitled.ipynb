```python
import json
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
```

```python
with open('/mnt/Enterprise/miccai_2023_CRIS/CRIS.pytorch/datasets/anns/refcoco/train.json') as json_file:
    prompts_list = json.load(json_file)
```

```python
sentences_all = []
for prompts in prompts_list:
    sentences = prompts['sentences']
    for sentence in sentences:
        one_sentence = sentence['sent']
        sentences_all.append(one_sentence)
```

```python
len(sentences_all)
```

```python
sentences_all[0]
```

```python
sentences_all
```

```python
words_all = " ".join(sentences_all)
```

```python
len(words_all.split(" "))
```

```python
words_frequency = Counter(words_all.split(" "))
```

```python
words_frequency
```

```python
words_frequency = dict(sorted(words_frequency.items(), key=lambda item: item[1], reverse=True))
```

```python
words_frequency
```

```python
list(words_frequency.keys())[:70]
```

```python
key, val = list(words_frequency.keys())[:19], list(words_frequency.values())[:19]
```

```python
plt.figure()
plt.bar(key, val)
plt.title('Words frequency in refcoco annotation used during traing CRIS')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=60)
plt.grid(linestyle='--')
```
